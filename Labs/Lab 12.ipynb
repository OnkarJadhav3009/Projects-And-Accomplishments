```python
import pandas as pd
import matplotlib.pyplot as plt
numerator = 366
denominator = 366
probabilities = []

for i in range(2, 100):
    numerator *= (366 + 1 - i)
    denominator *= 366
    probabilities.append({'group_size':i, 'probability':round(1 - (numerator / denominator), 3)})


df = pd.DataFrame(probabilities)

x_line = []
y_line = []
for i in df['group_size']:
    x_line.append(i)

for i in df['probability']:
    y_line.append(i)
plt.figure(figsize=(12, 4))
plt.scatter(x_line, y_line)
plt.title("Probability of sharing birthday on same day")
plt.xlabel("Size of the group")
plt.ylabel("Probabilities")
plt.show()
```
